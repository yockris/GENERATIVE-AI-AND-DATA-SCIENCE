# Import all the necessary libraries

- Matplotlib
- Pandas
- Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')
```

## Read the Dataframe using .csv

```python
cars = pd.read_csv('cars_data.csv')
```

## Show the first 5 Rows of the csv file using .head

```python
cars.head()
```

# Create a basic LinePlot using Matplotlib

- Read the first 30 Rows from the Dataframe
- Label X-axis for the Index; first 30 Rows of the data from the Dataframe
- Label Y-axis to read each column of the Dataframe

```python
plt.figure(figsize=(8,4))
cars.head(30).reset_index(drop=True).plot(marker='o')
plt.title('Car Sales Data')
plt.xlabel('Index')
plt.ylabel('Cars_Data')
plt.show()
```

# Create a ScatterPlot using Matplotlib

- Plot Cars(dataframe) Model and Mileage
- Label X-axis to read the Cars(dataframe) Model
- Label Y-axis to read the Cars(dataframe) Mileage

```python
plt.figure(figsize=(14,7))
plt.scatter(cars.head(30)['Model'], cars.head(30)['Mileage'], alpha=0.6)
plt.title('Model vs Mileage')
plt.xlabel('Model')
plt.ylabel('Mileage')
plt.show()
```

# Visualize the Cars Dataframe with Histogram

```python
plt.figure(figsize=(14,7))
sns.histplot(cars['Price'], bins=20, kde=True).set_title('Car Sales Data')
plt.show()
```

# Create a BoxPlot using Seaborn

```python
plt.figure(figsize=(14,7))
sns.boxplot(x='Fuel type', y='Mileage', data=cars).set_title('Mileage by Fuel Type')
plt.show()
```
